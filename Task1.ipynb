```python
category_sales = df.groupby('Category')['Sales'].sum().reset_index()
top_category = category_sales.loc[category_sales['Sales'].idxmax()]
print(f"Highest revenue category: {top_category['Category']}")
print(f"sales amount ${top_category['Sales']}")
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=category_sales, x='Category', y='Sales', palette='viridis')
plt.title('Revenue by Product Category')
plt.xlabel('Category')
plt.ylabel('Sales($)')
plt.show()
```
